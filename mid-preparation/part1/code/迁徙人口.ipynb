```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
path = '../data/population.csv'
data = pd.read_csv(path)
#data.set_index(['date'],inplace=True)
data.head()
```

```python
x = pd.to_datetime(data.date,format = "%Y%m%d")
y1 = data.今年
y2 = data.去年
font = {'family': 'STSong', 'weight': 'normal', 'size': 20}
```

```python
plt.figure(figsize = (20, 10))
plt.plot(x,y1,'r-',label = '今年迁徙人口',linewidth = 5.0)
plt.plot(x,y2,'b--',label = '去年农历同期迁徙人口',linewidth = 5.0)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
matplotlib.rcParams['font.family'] = 'STSong'
plt.title("武汉迁出趋势图",fontsize=30)
plt.legend(loc=1,prop = font)
```

```python
plt.figure(figsize = (20, 10))
#x1 = data.loc[0:14, ['date']]
x1 = pd.to_datetime(data.date.loc[0:14],format = "%Y%m%d")
z1 = data.loc[0:14, ['今年']]
z2 = data.loc[0:14, ['去年']]
plt.plot(x1,z1,'ro-',label = '今年迁徙人口',linewidth = 5.0,ms = 15)
plt.plot(x1,z2,'b*--',label = '去年农历同期迁徙人口',linewidth = 5.0, ms =15)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
matplotlib.rcParams['font.family'] = 'STSong'
plt.title("武汉迁出趋势图(部分)",fontsize=25)
plt.legend(loc=2,prop = font)
```
